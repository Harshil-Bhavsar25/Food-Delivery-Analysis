```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv('food_delivery_data.csv')
df
```

```python
# description of columns

# ID: String (object). Unique identifier for each delivery record, formatted as a hexadecimal string (e.g., `0xd224`).
# Delivery_person_ID: String (object). Unique identifier for the delivery person, combining city code and a specific ID (e.g., `DEHRES11DEL03`).
# Delivery_person_Age: Float. Age of the delivery person, ranging from 20 to 39 years, with some missing values.
# Delivery_person_Ratings: Float. Rating of the delivery person, ranging from 3.2 to 5.0, with some missing values.
# Restaurant_latitude: Float. Latitude coordinate of the restaurant, with some entries as 0.0 indicating missing or invalid data.
# Restaurant_longitude: Float. Longitude coordinate of the restaurant, with some entries as 0.0 indicating missing or invalid data.
# Delivery_location_latitude: Float. Latitude coordinate of the delivery location, with small increments from restaurant latitude.
# Delivery_location_longitude: Float. Longitude coordinate of the delivery location, with small increments from restaurant longitude.
# Order_Date: String (object). Date of the order in DD-MM-YYYY format (e.g., `15-02-2022`).
# Time_Ordered: String (object). Time when the order was placed in HH:MM:SS format (e.g., `21:35:00`), with some missing values.
# Time_Order_picked: String (object). Time when the order was picked up in HH:MM:SS format (e.g., `21:45:00`).
# Weatherconditions: String (object). Weather condition during delivery, with categories like Sunny, Cloudy, Fog, Windy, Stormy, Sandstorms.
# Road_traffic_density: String (object). Traffic density during delivery, categorized as Low, Medium, High, or Jam.
# Vehicle_condition: Float. Condition of the delivery vehicle, ranging from 0.0 to 2.0, representing poor to good condition.
# Type_of_order: String (object). Type of food order, with categories like Buffet, Drinks, Meal, Snack.
# Type_of_vehicle: String (object). Type of vehicle used for delivery, with categories like motorcycle, scooter, electric_scooter.
# multiple_deliveries: Float. Number of deliveries in a single trip, ranging from 0.0 to 2.0, with some missing values.
# Festival: String (object). Indicates if the delivery occurred during a festival, with values `Yes` or `No`.
# City: String (object). Type of city where delivery occurred, with categories Urban, Metropolitian, Semi-Urban.
# Time_taken(min): Float. Time taken for delivery in minutes, ranging from 10.0 to 54.0.
```
